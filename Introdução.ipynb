```python
def sigrms(sinal): #código para encontrar o sinal RMS
    c=sinal.size
    ret=np.zeros([c])
    for i in range (c):
        quad=sinal[i]**2
        raiz=quad** (1/2)
        ret[i]=raiz
    return ret
```

```python
from IPython.display import Image, HTML
import numpy as np      
import matplotlib.pyplot as plt
import imutils
from scipy import signal

def convhann(sinal,kernel):
    win2d = np.ones([kernel,kernel])
    win = signal.hann(kernel)
    np.sqrt(np.outer(np.ones(kernel),win,win2d)) 
    filtered = signal.convolve(sinal, win2d, mode='same') 
    return filtered
url1="https://upload.wikimedia.org/wikipedia/commons/thumb/1/11/1987_listrada_coca_cola_vermelho.png/636px-1987_listrada_coca_cola_vermelho.png"
url2="https://petervan.files.wordpress.com/2012/12/esher.png"
#importa as duas imagens da internet em escala de cinza
img0=imutils.url_to_image(url2,0)
img1=imutils.url_to_image(url1,0)
#faz a convolução
img4=convhann(img0,8)
img5=convhann(img1,8)
plt.figure()
plt.figure(figsize=(25,25))
plt.subplot(321),plt.imshow(img0,"gray"),plt.title('imagem original')
plt.subplot(322),plt.imshow(img1,"gray"),plt.title('imagem original')
plt.subplot(323),plt.imshow(img4,"gray"),plt.title('imagem média hann')
plt.subplot(324),plt.imshow(img5,"gray"),plt.title('imagem média hann')
plt.show()
```

```python
from IPython.display import Image, HTML
import numpy as np      
import matplotlib.pyplot as plt
import imutils
from scipy import signal

def convdif(sinal,kernel):
    c=sinal.size
    ret=np.zeros([c])
    win = np.ones([kernel,kernel])
    win[int(kernel/2):int(kernel)]=-1
    ret = signal.convolve(sinal, win, mode='same') 
    return ret
url1="https://upload.wikimedia.org/wikipedia/commons/thumb/1/11/1987_listrada_coca_cola_vermelho.png/636px-1987_listrada_coca_cola_vermelho.png"
url2="https://petervan.files.wordpress.com/2012/12/esher.png"
#importa as duas imagens da internet em escala de cinza
img0=imutils.url_to_image(url2,0)
img1=imutils.url_to_image(url1,0)
#faz a convolução
img4=convdif(img0,8)
img5=convdif(img1,8)
plt.figure()
plt.figure(figsize=(25,25))
plt.subplot(321),plt.imshow(img0,"gray"),plt.title('imagem original')
plt.subplot(322),plt.imshow(img1,"gray"),plt.title('imagem original')
plt.subplot(323),plt.imshow(img4,"gray"),plt.title('imagem filtro de sobel 1d')
plt.subplot(324),plt.imshow(img5,"gray"),plt.title('imagem filtro de sobel 1d')
plt.show()
```

```python
from IPython.display import Image, HTML
import numpy as np      
import matplotlib.pyplot as plt
import imutils
from scipy import signal

def limiarmais(sinal,valor):
    c,l=sinal.shape
    ret=np.zeros([c,l])
    for i in range (c):
        for j in range (l):
            if(sinal[i,j]>valor):
                ret[i,j]=sinal[i,j]
    return ret
url1="https://upload.wikimedia.org/wikipedia/commons/thumb/1/11/1987_listrada_coca_cola_vermelho.png/636px-1987_listrada_coca_cola_vermelho.png"
url2="https://petervan.files.wordpress.com/2012/12/esher.png"
#importa as duas imagens da internet em escala de cinza
img0=imutils.url_to_image(url2,0)
img1=imutils.url_to_image(url1,0)
#faz a convolução
img4=limiarmais(img0,160)
img5=limiarmais(img1,160)
plt.figure()
plt.figure(figsize=(25,25))
plt.subplot(321),plt.imshow(img0,"gray"),plt.title('imagem original')
plt.subplot(322),plt.imshow(img1,"gray"),plt.title('imagem original')
plt.subplot(323),plt.imshow(img4,"gray"),plt.title('imagem limiarizada')
plt.subplot(324),plt.imshow(img5,"gray"),plt.title('imagem limiarizada')
plt.show()
```
